```python
gu_mapping = {
    '도봉구': 1, '강북구': 2, '은평구': 3, '중랑구': 4, '양천구': 5, '성북구': 6, '서대문구': 7, '노원구': 8,
    '동대문구': 9, '금천구': 10, '성동구': 11, '동작구': 12, '강서구': 13, '관악구': 14, '구로구': 15,
    '광진구': 16, '강동구': 17, '용산구': 18, '영등포구': 19, '마포구': 20, '송파구': 21, '종로구': 22,
    '서초구': 23, '중구': 24, '강남구': 25
}
```

```python
# '시군구명' 칼럼의 값을 매핑된 번호로 변경하는 함수
def map_to_number(name):
    return gu_mapping[name]
```

```python
# '시군구명' 칼럼에 apply 함수 적용하여 번호로 변경
df['구_번호'] = df['시군구명'].apply(map_to_number)
```
